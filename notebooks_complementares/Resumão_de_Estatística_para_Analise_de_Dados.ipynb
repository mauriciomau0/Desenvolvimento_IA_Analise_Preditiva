# Resumão - Estatística

**Variáveis podem se classificar em:**
* Quantitativas: Quando podem ser medidas numericamente. (ex: idade, altura e peso)
* Qualitativas: Quando não podem ser medidas numericamente. (ex: etnia, gênero, naturalidade)
* Variáveis unidimensionais: Somente transmitem informação sobre uma característica. (ex: idade).
* Variáveis bidimensionais: Guardam informação sobre duas características da população. (ex: idade e renda).
* Variáveis pluridimensionais: Detém informação sobre três ou mais características. (ex: idade, altura e peso).

**Variáveis quantitativas podem ser:**
* Discretas: Só podem tomar valores inteiros como o número azeitonas em uma pizza. Ex (1, 5, 6, -2)
* Contínuas: Podem tomar qualquer valor real dentro de um intervalo. Por exemplo, a velocidade de um veículo , que pode ser 70,4 km/h, 94,57 km/h...

**Conceitos fundamentais:**
* Indivíduo: Qualquer elemento que porte informação sobre o fenômeno que se estuda.
Assim, se estudamos à altura dos alunos em uma turma, cada aluno é um indivíduo;
se o objeto de estudo é o preço de uma habitação, cada unidade de habitação é um indivíduo.
* População: Conjunto de todos os indivíduos (pessoas, objetos, animais etc) que portem informação sobre o fenômeno que se estuda. Por exemplo, se estudamos o preço de habitações em uma cidade, a população será o total de habitações em dita cidade.
* Amostra: Subconjunto que, selecionando uma população, por exemplo, no caso do estudo do preço de habitações em uma cidade, o normal seria não obter informações sobre todas as moradias da cidade, pois seria um trabalho muito complexo, mas deveria selecionar um subgrupo (amostra) que se entenda suficientemente representativo.
* Rol: É a ordenação numérica dos dados em ordem não crescente ou não decrescente, para melhor visualização. (ex: 1, 5, 6, 6, 8, 8, 15)

* Classes: São a quantidade de grupamentos que faremos dos dados em Rol.
  - Usualmente, utilizamos: **K=√n**, onde **"n"** é o número de termos da distribuição estatística.
  * **K** deve ser um número inteiro, sendo recomendado o uso do mesmo entre 5 e 20.
  * **K** pode atender outros critérios estabelecidos pelo analista.

* Amplitude
  * Amplitude total **(A)** da distribuição de dados é a diferença entre o maior e o menor termo da distribuição.
  * Amplitude do intervalo de classes **(c)** se da por: **(max-min)/(k-1)**

```python
# Portanto, em uma distribuição de 50 dados discretos aleatórios situados entre 0 e 100:
import random
random.seed(42)

#Determinando a amostra
sample = random.sample(range(1, 100), 50)
print("A amostra é:", sample)
print()

#Determinando o Rol da amostra
rol = sorted(sample)
print("O rol é:", rol)
print()

#Determinando o número de elementos na amostra
n = len(sample)
print("O número de elementos 'n' é:", n)
print()

#Determinando o maior valor da amostra
max_value = max(sample)
print("O maior valor é", max_value)
print()

#Determinando o menor valor da amostra
min_value = min(sample)
print("O menor valor é", min_value)
print()

#Determinando o número de classes
k = int(n**(1/2))
print("O número de classes 'k' é:", k)
print()

#Determinando a amplitude total da amostra
A = max_value - min_value
print("A amplitude total 'A' é:", A)
print()

#Determinando a amplitude do intervalo de classes classe da amostra
c = (max_value - min_value)/(k-1)
print("A amplitude do intervalo de classes 'c' é:", round(c,3))
```

  **Intervalos de classe**.
* Limite inferior **(LI)**
  * O menor valor encontrado em uma classe.
* Limite superior **(LS)**
  * O maior valor encontrado em uma classe.
* Intervalos abertos
  * Quando os limites da classe (inferior e superior) não pertencem a ela.
* Intervalos fechados
  * Quando os limites de classe (superior e inferior) pertencem à classe em questão.
* Intervalos mistos
  * Quando um dos limites pertence à classe, e o outro, não.

**Frequência**
* A frequência absoluta **(f)**
  * Corresponde ao número de observações que temos em uma determinada classe.
* A frequência relativa **(fr)**
  * Corresponde à proporção do número de observações em uma determinada classe, em relação ao total de observações.
* A frequência acumulada **(fa)**
  * Corresponde ao número de observações abaixo do limite inferior de uma determinada classe.
* A frequência acumulada relativa **(fa.r)**
  * Corresponde à proporção do número de observações abaixo do limite inferior de uma determinada classe, em relação ao total de observações.

```python
#Portanto, na utilização de dados contínuos referentes a variação proporcional diária do preço do barril de petróleo
import pandas as pd

#Importando os dados da amostra
df = pd.read_csv(f"https://drive.google.com/uc?export=download&id={'1l3ocWx0iNHiEdxa_YVSYFHuYd7VPtEtn'}")
sample = df.iloc[:, 0].tolist()
#Determinando o rol da amostra
rol = sorted(sample)
#Determinando o número de elementos na amostra
n = len(sample)
#Determinando o maior valor da amostra
max_value = max(sample)
#Determinando o menor valor da amostra
min_value = min(sample)
#Determinando o número de classes
k = int(n**(1/2))
#Determinando a amplitude total da amostra
A = max_value - min_value
#Determinando a amplitude do intervalo de classes classe da amostra
c = (max_value - min_value)/(k-1)

#Criando as classes
LI = [min_value + i*c for i in range(k)]
LS = [min_value + (i+1)*c for i in range(k)]
LS[-1] = max_value  # Ajuste para incluir o valor máximo na última classe

# Exibindo a tabela de intervalos e frequências
classes = [f"{i+1}ª Classe" for i in range(k)]
F = [sum(1 for x in rol if LI[i] <= x < LS[i]) for i in range(k)]
Fa = [sum(F[:i+1]) for i in range(k)]

classes_df = pd.DataFrame({
    'LI': LI,
    'LS': LS,
    'F': F,
    'Fr': [f/n for f in F],
    'Fa': Fa,
    'Fa.r': [fa/n for fa in Fa]
}, index=classes)

print(classes_df)
```

**Medidas de posição**
  * Constituem uma forma mais sintética de apresentar os resultados contidos nos dados observados, pois representam um valor central.
* Média aritmética
  * Soma de todos os termos dividido pelo número de termos.
* Média ponderada
  * Cada valor, ou conjunto de valores multiplicado ao seu peso e dividido pelo total dos pesos.
* Média geométrica
  * Todos os termos multiplicados, fatorados pela raiz de **"n"**, sendo **"n"** o número de termos.
* Mediana
  * Se número de termos for impar, termo central. Se par, a média dos termos centrais.
* Moda
  * Termo mais frequente

```python
#Exemplos para um conjunto de 50 variáveis discretas:
import random
import statistics
import math
random.seed(42)

#Determinando a amostra
sample = random.sample(range(1, 100), 50)
n = len(sample)
print("A amostra é:", sample)
print()

#Calculando a média aritmética
media = sum(sample)/n
print("A média aritmética é:", media)
print()

#Calculando a média ponderada
media_ponderada = sum(sample[i]*10 for i in range(n))/n
print("A média ponderada para um peso fixo de 10 é:", media_ponderada)
print()

#Calculando a média geometrica
media_geometrica = (math.prod(i for i in sample))**(1/n)
print("A Média Geométrica:", round(media_geometrica,3))
print()

#Calculando a mediana
mediana = statistics.median(sample)
print("A Mediana é:", mediana)
print()

#Calculando a moda
moda = statistics.mode(sample)
print("A Moda é:", moda)
```

A mediana é menos sensível a "outliers" que a média, portanto é uma melhor escolha para dados simétricos. Muito cuidado com sua utilização em distribuições de dados assimétricos.

**Medidas de dispersão ou variabilidade**

* Variância
  * Medida da sua dispersão estatística, indicando "o quão longe", em geral, os seus valores se encontram do valor esperado: **(Σ(Xi-média)² )/n**, Sendo: Xi = Cada termo, Σ = Soma, n = Número de termos
* Desvio padrão
  * Raiz quadrada da variância.
* Coeficiente de Variação **(CV)**
  * Consiste em uma medida relativa de dispersão, útil para a comparação em termos relativos ao grau de concentração em torno da média de séries distintas, descrita por: **Desvio Padrão/Média**
  * Baixa dispersão: CV ≤ 0,15
  * Média dispersão: 15% < CV < 0,3
  * Grande dispersão: CV ≥ 0,3
  * O Coeficiente de variação também pode se apresentar em forma de porcentagem

```python
#Exemplos para um conjunto de 50 variáveis discretas:
import random
random.seed(42)

#Determinado a amostra
sample = random.sample(range(1, 100), 50)
print("A amostra é:", sample)
print()

#Calculando a variancia
media = sum(sample)/n
variancia = sum([(i-media)**2 for i in sample])/(n-1)
print("A Variância é:", round(variancia,3))
print()

#Calculando o desvio padrão
desvio_padrao = variancia**(1/2)
print("O Desvio Padrão é:", round(desvio_padrao,3))
print()

#Calculando o coeficiente de variação
cv = (desvio_padrao/media)*100
print(f"O Coeficiente de Variação é: {round(cv, 3)}%")
```

**Probabilidades:**
* É a ferramenta utilizada para estabelecer relação entre uma amostra e sua população.
* Inicialmente, o resultado é desconhecido, mas apresenta uma certa regularidade ao longo do tempo.
* Geralmente, as probabilidades são representadas por frações, mas também podem ser expressas em porcentagens ou números decimais.
* Qualquer probabilidade deve estar necessariamente entre 0 e 1: **0 < P(e) < 1**.

**Conceitos fundamentais:**
* Experimento Aleatório
  * Situações as quais o resultado a cada execução é aleatório, como ao jogar uma moeda para cima ou um dado sobre uma mesa.
* Espaço Amostral
  * É o conjunto de elementos as quais pertencem todas os possíveis resultados, representado pela letra **"S"**. Por exemplo, ao lançar um dado, o espaço amostral seria **"S = {1, 2, 3, 4, 5, 6}"**.
* Evento
  * Quando lançamos um dado ou uma moeda, chamamos a ocorrência deste fato de evento.
  * Todo subconjunto de um espaço amostral é considerado um evento.
  * Em relação ao espaço amostral de um lançamento de um dado, o conjunto **"A= {2,3,5}"** é um subconjunto de **"S"**, que representa o resultado do lançamento quando a face para cima exibe um número primo.

**Classificação de Eventos:**
* Evento Simples
  * São eventos formados por um único elemento do espaço amostral. Por exemplo, obter o resultado **"A = {cara}"** ao lançar uma moeda.
* Evento Certo
  * É um evento que comporta todos os resultados possíveis no espaço amostral, assim como, ao lançar um dado, obter um número que seja divisor de 720. Já que 6! = 6x5x5x3x2x1 = 720, esse evento pode ser representado por **A = {1,2,3,4,5,6}**.
* Evento Impossível
  * É um evento que não existe possibilidade de acontecer, como, ao lançar dois dados, a soma das duas faces ser maior que 12. Podemos representá-lo por **"A = {?}"**.
* Evento União
  * Considerando **A = {1, 2}** como a representação de um evento em que o resultado do lançamento de um dado é inferior a 3 e **B = {6}** como a representação do resultado superior a 5, o evento união será a **C = {1, 2, 6}**, uma vez que **C = A ∪ B**. Nesse caso, **"C"** contém todos os elementos de **"A"** e **"B"**.
* Evento Intersecção
  * Considerando **A = {2, 4, 6}** como a representação de um evento em que o resultado do lançamento de um dado par e **B = {6}** como a representação do resultado superior a 5, o evento intersecção será a **C = {6}** uma vez que **C = A ∩ B**. Nesse caso, **"C"** contém todos os comuns de **"A"** e **"B"**.
* Eventos Mutuamente Exclusivos
  * São eventos que não têm interseções entre seus conjuntos. Considerando **A = {5}** como a representação de um evento que o resultado do laçamento de um dado é divisível por 5 e **B = {1,3}** a representação do resultado ser impar e menor ou igual a 3, os eventos **"A"** e **"B"** são considerados mutuamente exclusivos, uma vez que **A ∩ B = ∅**.
* Evento Complementar
  * Considerando **A = {2, 4, 6}** como a representação de um evento em que o resultado do lançamento de um dado é um número par, seu evento complementar é **Ā = {1, 3, 5}**, que representa os resultados ímpares. Os elementos de **Ā** são todos que pertencem ao conjunto **S** e que estão presentes **A**. Portanto, **Ā = S - A**, ou **S = A + Ā**.

**Probabilidade de ocorrência de um evento:**
* Definição de Probabilidade
  * Probabilidades são valores entre 0 e 1 que mostram a chance de acontecimento de um evento, denotado por **"P"**.
  * As probabilidades podem se apresentar de diversas maneiras como frações, porcentagens, números decimais e entre outros.
* Fenômenos Determinísticos
  * São aqueles os quais os resultados sempre serão os mesmos, independente do número de ocorrências, ou seja, possuem probabilidade igual a 1.
* Fenômenos Aleatórios
  * São aqueles os quais os resultados são imprevisíveis a cada ocorrência, mesmo que haja inúmeras repetições, ou seja, possuem probabilidade igual a 0.
* Probabilidade no Caso de um Evento Simples
  * Para um evento **E** qualquer, será dada por **"P(E) = n(E) / n(S)"**, para **n(S) ≠ 0**, onde **"n(E)"** representa o número de elementos do conjunto evento e **"n(S)"** o número de elementos do espaço amostral.
* Probabilidade no Caso da União de Eventos
  * Para dois eventos **"A"** e **"B"** quaisquer será dada por **"P(A∪B) = P(A) + P(B) - P(A∩B)"**.
* Probabilidade no Caso de Eventos Mutuamente Exclusivos
  * Considerando **A ∩ B ≠ 0**, esses eventos serão mutualmente exclusivos, pois **P(A∩B) = 0**. Considerando **A ∩ B ≠ 0**, esses eventos serão mutualmente exclusivos, pois **P(A∩B) = 0**.
* Caso da Probabilidade Condicionada
  * Considerando **"A"** e **"B"** dois eventos quaisquer de um espaço amostral **"S"** finito e não vazio. A probabilidade de **B** condicionada a **A** é dada pela probabilidade de ocorrência de **B**, sabendo que já ocorreu **A**. É representada por P(B/A).
* Situação da Intersecção de Eventos
  * Para dois eventos quaisquer, **"A"** e **"B"**, em um espaço amostral **"S"** que é finito e não vazio, a relação é dada por **"P(A∩B) = P(A) × P(B)"**.

```python
#Usando o experimento aleatório do lançamento de um dado:
import random
random.seed(42)

#Definindo o espaço amostral de cada evento.
S = [1,2,3,4,5,6]
print("O espaço amostral 'S' para o lançamento de um dado é:", S)
print()

#Definindo um evento simples "A", no qual a face virada para cima de um dado após ser lançado é par.
A = [i for i in S if i % 2 == 0]
P_A = len(A)/len(S)
print(f'A probabilidade da face para cima de um dado após ser lançado ser par, dado pelo evento simples "A = {A}", é de {round(P_A*100)}%.')
print()

#Definindo um evento união "B", no qual a face virada para cima de um dado após ser lançado é par ou maior que 3.
B = [i for i in S if i % 2 == 0 or i > 3]
P_B = len(B)/len(S)
print(f'A probabilidade da face para cima de um dado após ser lançado ser par ou maior que 3, dado pelo evento união "B = {B}", é de {round(P_B*100,2)}%.')
print()

#Definindo dois eventos mutualmente exclusivos em que, a face virada para cima após ser lançado 2 e a face virada para cima de um dado após ser lançado é maior que 5, respectivamente.
C = [i for i in S if i < 2 and i % 5 == 0]
P_C = len(C)/len(S)
print(f'A probabilidade da face para cima de um dado após ser lançado ser menor que 2 e divisivel por 5, dado pelo evento "C = {C}", é igual a {round(P_C)}%, portanto esses são eventos mutualmente exclusivos.')
print()

#Definindo um evento intersecção "D", no qual a face virada para cima de um dado após ser lançado é par e maior que 3.
D = [i for i in S if i % 2 == 0 and i > 3]
P_D = len(D)/len(S)
print(f'A probabilidade da face para cima de um dado após ser lançado e maior que 3, dado pelo evento intersecção "D = {D}", é de {round(P_D*100,2)}%.')
```

**Tipos de Amostragem:**
* Amostragem Não Probabilística
  * Utiliza-se geralmente a amostragem não probabilística pela simplicidade intrínseca ao processo, bem como pela impossibilidade de se obter uma amostra probabilística, como seria desejável. Nesse tipo de amostragem não conhecemos a priori a probabilidade que um elemento da população tem de pertencer à amostra.
* Amostragem a Esmo
  * É uma forma de amostragem não probabilística, onde selecionamos elementos aleatoriamente para compor a amostra, sem seguir qualquer critério específico. Quanto mais homogênea for a população, maior será a possibilidade de considerarmos que a amostra é representativa da população.
* Intencional
  * Trata-se de um tipo de amostra em que os indivíduos de uma população são selecionados intencionalmente com base em critérios previamente definidos. Geralmente, esse método é aplicado em estudos voltados para um público-alvo específico, como mulheres de classe social menor que B, por exemplo.
* Quotas ou proporcional
  * Esse tipo de amostra refere-se a um subgrupo da população que atende a critérios específicos, preservando as proporções de outras variáveis entre os elementos. Assim, os indivíduos não são escolhidos aleatoriamente, mas selecionados até que o tamanho dos grupos desejados seja atingido. Por exemplo, pode-se ter uma amostra de pessoas da classe A, respeitando a proporção de homens e mulheres existentes na população geral.
* Desproporcional
  * Utilizada quando a escolha da amostra for desproporcional à população. Atribuem-se pesos para os dados, e assim obtém-se resultados ponderados representativos para o estudo.
* Probabilística
  * Esse é o método mais indicado para realizar inferências sobre a população, pois todos os elementos da população possuem a mesma chance de serem incluídos na amostra.
* Aleatória Simples
  * Os indivíduos são escolhidos para a amostra por meio de um sorteio simples, garantindo que todos tenham a mesma probabilidade de serem selecionados. Essa metodologia de amostragem deve ser aplicada apenas quando a população for homogênea em relação à variável que se deseja estudar. Não é recomendado quando os dados estão ordenados.
* Amostragem Sistemática
  * Os indivíduos que compõem a amostra são selecionados de forma cíclica, o que é interessante para preservar a aleatoriedade dos dados quando estes se apresentam organizados. Após definir o valor de **"K"**, que representa a quantidade de classes, esse valor funcionará como um fator cíclico. Ao escolher um elemento do primeiro intervalo, os próximos serão escolhidos seguindo a fórmula **(i+K)**, sempre somando **"K"** ao último elemento já incluído na amostra, até que o número de classes seja alcançado.
* Amostragem Aleatória Estratificada
  * A amostragem estratificada é utilizada quando a população apresenta heterogeneidade e a variável de interesse também demonstra essa característica. Dessa forma, são criados subgrupos que são homogêneos entre si em relação a outras variáveis, como etnia, renda, sexo e idade, entre outras. É fundamental que a amostragem estratificada mantenha a proporção desses grupos homogêneos dentro da população heterogênea, assegurando que a representação de cada subgrupo na amostra final seja precisa e reflita corretamente a população como um todo.
* Conglomerado
  * A amostragem por conglomerado é uma opção mais simples em comparação à amostragem estratificada. Essa abordagem diminui consideravelmente os custos de processamento da amostra. Um conglomerado é um subgrupo da população que reflete suas características, ou seja, seus elementos são igualmente heterogêneos entre si. Para definir conglomerados, podem ser utilizadas famílias, empresas ou qualquer outra entidade que reúna indivíduos.

**Dimensionamento das Amostras:**
  * Um dos aspectos mais relevantes em uma pesquisa estatística é determinar o tamanho adequado das amostras. Amostras excessivamente grandes podem ser custosas e difíceis de gerenciar, enquanto amostras muito pequenas tendem a ser imprecisas e pouco confiáveis.

* O processo deve se basear em três etapas:
  * Avaliar a variável mais importante do grupo e a mais significativa.
  * Analisar se é ordinal, intervalar ou nominal.
  * Verificar se a população é finita ou infinita.

* Algumas expressões propostas para determinar o tamanho da amostra em alguns casos comuns:
  * Variável intervalar e população infinita: **n = ((Z.σ)/N)²**
  * Variável intervalar e população finita: **n = (Z².σ²N)/(d²(N-1)+Z².σ²)**
  * Variável nominal ou ordinal e população infinita: **n = (Z².pq)/d²**
  * Variável nominal ou ordinal e população finita: **n = (Z².p.q.N)/(d²(N-1)+Z².p.q)**
  * Em que **"d"** representa o erro admissível, **"Z"** é uma função relacionada ao nível de confiança, **"N"** refere-se ao tamanho da população e **"p"** indica a proporção da variável considerada, então **"q"** será o complemento, ou seja, **"q = 1-p"**. Quando as proporções são desconhecidas, adota-se **"p = q = 0,5"**. Além disso, como já é sabido, **"σ"** denota o desvio padrão.

```python
#Portanto, tomando como dado, "Z" = 1,95, "d" = 400 E "σ" = 1200
Z = 1.95
d = 400
σ = 1200

#Calculando o número de amostras:
n = ((Z*σ)/d)**2
print(f'O número de amostras é de {round(n)}.')
```
